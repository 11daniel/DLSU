## Introduction to the Task and Dataset

The chosen task for the model is to predict the `Per Capita Income` of a certain family given some features

## Description of the Dataset

### Brief Description of the Dataset

The dataset comes from the Philippine Statistics Authority. It contains data about expenditures and other disbursements, housing characteristics, social protection, Income and other receipts and Entrepreneural Activities.

### Data Collection Process
The dataset was collected through two face-to-face survey visits, conducted in 2012 and 2013, with data covering two phases of the year to capture seasonal variations in income and expenditure. The process was closely supervised by regional and provincial staff from the Philippine Statistics Authority (PSA) to ensure accurate data collection, with supervisors conducting random re-interviews and checking the quality of enumerator work. Trained enumerators collected detailed information on family income sources and expenditures. While the thorough supervision and structured approach likely ensured data reliability, challenges such as non-response, deferment of interview, lack of forms, problem area, and accidents or injuries could have impacted data completeness. The researchers, however, did have protocols for each of these potential enumeration and related problems. The method’s strength lies in its ability to capture nuanced, seasonal financial patterns, though potential biases from interviewees' reluctance to disclose sensitive financial information must be considered when interpreting the dataset.

### Structure of the Dataset
The dataset is composed of 40171 rows (instances) and 119 columns (features). Each row represents a case while each column represents a variable.

```python
from sklearn.metrics import make_scorer, mean_squared_error, r2_score
from sklearn.linear_model import Ridge
from sklearn.linear_model import SGDRegressor
import numpy as np

# Define the parameter grid for Ridge
param_grid_ridge = {
    "alpha": [0.01, 0.1, 1, 10, 100]  # Range of regularization strengths
}

ridge = Ridge()
mse_scorer = make_scorer(mean_squared_error, greater_is_better=False)

ridge_grid_search = GridSearchCV(estimator=ridge, param_grid=param_grid_ridge, scoring=mse_scorer, cv=5)
ridge_grid_search.fit(X_train, y_train)

# Extract best Ridge model
best_ridge = ridge_grid_search.best_estimator_
best_ridge_params = ridge_grid_search.best_params_

ridge_y_pred_test = best_ridge.predict(X_test)
ridge_y_pred_train = best_ridge.predict(X_train)
ridge_mse_test = mean_squared_error(y_test, ridge_y_pred_test)
ridge_mse_train = mean_squared_error(y_train, ridge_y_pred_train)
ridge_r2_test = r2_score(y_test, ridge_y_pred_test)
ridge_r2_train = r2_score(y_train, ridge_y_pred_train)

print("Ridge Regression:")
print(f"Best Parameters: {best_ridge_params}")
print(f"Test MSE: {ridge_mse_test:.3f}")
print(f"Train MSE: {ridge_mse_train:.3f}")

# Stochastic Gradient Descent
sgd_model = SGDRegressor(
    learning_rate='constant',
    eta0=0.001,
    max_iter=500,
    penalty=None,
    random_state=42
)
sgd_model.fit(X_train, y_train)  # Fit SGD model

# SGD Predictions
sgd_y_pred_train = sgd_model.predict(X_train)
sgd_y_pred_test = sgd_model.predict(X_test)
sgd_mse_train = mean_squared_error(y_train, sgd_y_pred_train)
sgd_mse_test = mean_squared_error(y_test, sgd_y_pred_test)
sgd_r2_train = r2_score(y_train, sgd_y_pred_train)
sgd_r2_test = r2_score(y_test, sgd_y_pred_test)

print("\nStochastic Gradient Descent Regressor (SGD):")
print(f"Test MSE: {sgd_mse_test:.3f}")
print(f"Train MSE: {sgd_mse_train:.3f}")

# Batch Gradient Descent
def gradient_descent(X, y, learning_rate=0.1, num_iterations=1000):
    n_samples = len(y)
    w = np.zeros(X.shape[1])
    b = 0

    for i in range(num_iterations):
        y_pred = np.dot(X, w) + b
        error = y_pred - y

        dw = (1 / n_samples) * np.dot(X.T, error)
        db = (1 / n_samples) * np.sum(error)

        w -= learning_rate * dw
        b -= learning_rate * db

    return w, b

w_optimized, b_optimized = gradient_descent(X_train, y_train)


y_pred_train_batch = np.dot(X_train, w_optimized) + b_optimized
y_pred_test_batch = np.dot(X_test, w_optimized) + b_optimized
batch_mse_train = mean_squared_error(y_train, y_pred_train_batch)
batch_mse_test = mean_squared_error(y_test, y_pred_test_batch)
batch_r2_train = r2_score(y_train, y_pred_train_batch)
batch_r2_test = r2_score(y_test, y_pred_test_batch)


print("\nBatch Gradient Descent:")
print(f"Test MSE: {batch_mse_test:.3f}")
print(f"Train MSE: {batch_mse_train:.3f}")

if ridge_mse_test <= min(sgd_mse_test, batch_mse_test):
    final_model = "Ridge"
    final_mse_test = ridge_mse_test
    final_mse_train = ridge_mse_train
    final_r2_train = batch_r2_train
    final_r2_test = batch_r2_test

elif sgd_mse_test <= min(ridge_mse_test, batch_mse_test):
    final_model = "SGD Regressor"
    final_mse_test = sgd_mse_test
    final_mse_train = sgd_mse_train
    final_r2_train = batch_r2_train
    final_r2_test = batch_r2_test
else:
    final_model = "Batch Gradient Descent"
    final_mse_test = batch_mse_test
    final_mse_train = batch_mse_train
    final_r2_train = batch_r2_train
    final_r2_test = batch_r2_test

print(f"\nBest Model: {final_model}")
print(f"Best Model Test MSE: {final_mse_test:.3f}")
print(f"Best Model Train MSE: {final_mse_train:.3f}")
print(f"Best Model Test R²: {final_r2_test:.3f}")
print(f"Best Model Train R²: {final_r2_train:.3f}")
```

```python
def gradient_descent(X, y, learning_rate=0.1, num_iterations=1000):
    """
    Performs gradient descent to find the best coefficients for linear regression.

    Args:
        X: Training data features.
        y: Training data target variable.
        learning_rate: The learning rate for gradient descent.
        num_iterations: The number of iterations for gradient descent.

    Returns:
        w: The optimized coefficients for the linear regression model.
        b: The optimized intercept for the linear regression model.
        history: List of mean squared errors over all iterations.
    """
    n_samples = len(y)
    w = np.zeros(X.shape[1])
    b = 0
    history = []  # Track loss over iteration

    for i in range(num_iterations):
        y_pred = np.dot(X, w) + b
        error = y_pred - y
        mse = np.mean(error ** 2)
        history.append(mse)

        dw = (1 / n_samples) * np.dot(X.T, error)
        db = (1 / n_samples) * np.sum(error)

        w -= learning_rate * dw
        b -= learning_rate * db

    return w, b, history

# Perform gradient descent on training data
w_optimized, b_optimized, history_optimized = gradient_descent(X_train, y_train)

# Create a new linear regression model with the optimized coefficients
lin_reg_optimized = LinearRegression()
lin_reg_optimized.coef_ = w_optimized
lin_reg_optimized.intercept_ = b_optimized

# Make predictions using the optimized model
y_pred_train_optimized = X_train @ w_optimized + b_optimized
y_pred_test_optimized = X_test @ w_optimized + b_optimized

# Calculate R-squared for the optimized model
r2_train_optimized = r2_score(y_train, y_pred_train_optimized)
r2_test_optimized = r2_score(y_test, y_pred_test_optimized)
mse_train = mean_squared_error(y_train, y_pred_train_optimized)
rmse_train = root_mean_squared_error(y_train, y_pred_train_optimized)
mse_test = mean_squared_error(y_test, y_pred_test_optimized)
rmse_test = root_mean_squared_error(y_test, y_pred_test_optimized)

print(f"R² Score (Train, Optimized): {r2_train_optimized:.4f}")
print(f"R² Score (Testing, Optimized): {r2_test_optimized:.4f}")
print(f"Mean Squared Error (Train, Optimized): {mse_train:.4f}")
print(f"Mean Squared Error (Test, Optimized): {mse_test:.4f}")
print(f"Root Mean Squared Error (Train, Optimized): {rmse_train:.4f}")
print(f"Root Mean Squared Error (Test, Optimized): {rmse_test:.4f}")
```
